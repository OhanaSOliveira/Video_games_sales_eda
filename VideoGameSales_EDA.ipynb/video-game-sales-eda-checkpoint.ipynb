# Exploratory Analysis of the Video Game Sales Database 

In this notebook, we will explore the video game sales database, identifying sales patterns by year, genre, platform, and region.
Dataset imported from Kaggle: https://www.kaggle.com/datasets/sidtwr/videogames-sales-dataset/data

I begin the process by importing the Pandas, Matplotlib, and Seaborn libraries and adjusting the aesthetic settings of the future graphs.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações estéticas
sns.set(style="whitegrid")
%matplotlib inline
```

I read the dataset and look for the first 5 lines:

```python
df = pd.read_csv("vgsales.csv") 
df.head()
```

Now let's check the type of data we have in each column:

```python
df.info()
```

We then verified that we have 4 variables of type Object (Game Name(Name), Platform(Platform), Game Genre(Genre), Publisher(Publisher)), 1 integer variable (Sales Rank(Rank)), and 6 floating-point variables (Year of publication(Year), North American sales(NA_Sales), European sales(EU_Sales), Japanese sales(JP_Sales), sales in the rest of the world(Other_Sales), and global sales(Global_Sales)). We have 16598 rows and 11 columns in total. 

Next, we begin data processing by checking for empty spaces:

```python
df.isnull().sum()
```

We then verified that we have 271 missing values ​​in Year and 58 missing values ​​in Publisher. Therefore, I will replace the null values ​​in Publisher with 'Unknown'. Regarding the year, I considered the possibility of using the average, however, it is not viable, given that it does not reflect reality. I thought about deleting the rows, but this would cause the problem of excluding some games and bias our analysis. Therefore, I decided not to perform analyses using the year; perhaps this analysis would be interesting on a more complete dataset.

```python
df["Publisher"] = df["Publisher"].fillna("Unknown")
```

Now we will plot the top 10 best-selling games globally. I've defined the x-axis with the name of each game, and the y-axis with the number of global sales. I've set the color palette to Viridis for more intuitive coloring. 

```python
top10 = df.sort_values(by="Global_Sales", ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(data=top10, x="Name", y="Global_Sales", palette="viridis")
plt.title("Top 10 Best-Selling Games Globally")
plt.xticks(rotation=45)
plt.show()
```

Now let's see if this graph changes according to your region, starting with North America:

```python
top10 = df.sort_values(by="NA_Sales", ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(data=top10, x="Name", y="NA_Sales", palette="viridis")
plt.title("Top 10 Best-Selling Games in North America")
plt.xticks(rotation=45)
plt.show()
```

Comparing the two charts, we see that Wii Sports and Super Mario Bros. remain in the lead as the best-selling games. The rest change positions slightly.
Let's now analyze the situation in Europe:

```python
top10 = df.sort_values(by="EU_Sales", ascending=False).head(10) 
plt.figure(figsize=(12, 6))
sns.barplot(data=top10, x="Name", y="EU_Sales", palette="viridis")
plt.title("Top 10 Best-Selling Games in Europe")
plt.xticks(rotation=45)
plt.show()
```

Here we see a more drastic change and other titles appearing, such as NintendoDogs and Grand Theft Auto V. Let's summarize these graphs for comparison:


```python
#Region names
regions = {
    "North America": "NA_Sales",
    "Europe": "EU_Sales",
    "Japan": "JP_Sales",
    "Outhers": "Other_Sales",
    "Global": "Global_Sales"
}

# Layout subgrafic
fig, axs = plt.subplots(3, 2, figsize=(16, 15))
fig.suptitle("Top 5 Best-Selling Games by Region", fontsize=20)
axs = axs.flatten()

# Loop nas regiões
for i, (region_name, column) in enumerate(regions.items()):
    # 5 games most sales for region
    top_games = df[["Name", column]].sort_values(by=column, ascending=False).head(5)
    
    # Bar chart
    sns.barplot(
        data=top_games,
        x=column,
        y="Name",
        palette="viridis",
        ax=axs[i]
    )
    
    axs[i].set_title(region_name, fontsize=14)
    axs[i].set_xlabel("Sales (Millions)")
    axs[i].set_ylabel("Game Name")

# Remove gráfico vazio se sobrar (3x2 = 6 slots e temos 5 regiões)
if len(regions) < len(axs):
    axs[-1].axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

We see that there is a significant change in the top 5 for each region. Let's now see if this reflects regional tastes for each genre:

```python
# Paleta de cores personalizada para gêneros
genre_palette = {
    "Action": "red",
    "Sports": "blue",
    "Shooter": "purple",
    "Platform": "orange",
    "Racing": "green",
    "Misc": "pink",
    "Role-Playing": "gold",
    "Puzzle": "brown",
    "Simulation": "cyan",
    "Fighting": "gray",
    "Adventure": "teal",
    "Strategy": "olive"
}

# Regiões e colunas de vendas
regions = {
    "North America": "NA_Sales",
    "Europe": "EU_Sales",
    "Japan": "JP_Sales",
    "Others": "Other_Sales",
    "Global": "Global_Sales"
}

# Layout de subplots
fig, axs = plt.subplots(3, 2, figsize=(18, 16))
fig.suptitle("TTop 5 Bestselling Genres by Region", fontsize=20)
axs = axs.flatten()

for i, (region_name, column) in enumerate(regions.items()):
    # Soma de vendas por gênero
    genre_sales = df.groupby("Genre")[column].sum().sort_values(ascending=False).head(5).reset_index()

    # Gráfico de barras
    sns.barplot(
        data=genre_sales,
        x=column,
        y="Genre",
        palette=[genre_palette.get(genre, "gray") for genre in genre_sales["Genre"]],
        ax=axs[i]
    )

    axs[i].set_title(region_name, fontsize=14)
    axs[i].set_xlabel("Vendas (milhões)")
    axs[i].set_ylabel("Gênero")

# Remove plot extra, se houver
if len(regions) < len(axs):
    axs[-1].axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

```

We can see, then, that most regions have a preference for action games, except for Japan, which prefers role-playing games, explaining the top-selling games there.

Let's now see if the number of sales is impacted by the number of games published:

```python

# Agrupando e filtrando novamente (caso esteja em nova célula)
publisher_stats = df.groupby('Publisher').agg({
    'Name': 'count',
    'Global_Sales': 'sum'
}).rename(columns={'Name': 'Game_Count'})

publisher_stats = publisher_stats[publisher_stats['Game_Count'] > 1]
top_publishers_stats = publisher_stats.sort_values(by='Global_Sales', ascending=False).head(15)

# Criando gráfico com linha de regressão
plt.figure(figsize=(12, 8))
sns.regplot(
    data=top_publishers_stats,
    x='Game_Count',
    y='Global_Sales',
    scatter=True,
    ci=None,
    line_kws={"color": "red", "linewidth": 1.5}
)

# Adicionando os nomes das publishers
for publisher, row in top_publishers_stats.iterrows():
    plt.text(row['Game_Count'] + 0.3, row['Global_Sales'], publisher, fontsize=9)

plt.title('Top 15 Publishers: Number of Games vs. Global Sales (with trend line)')
plt.xlabel('Number of Games Published')
plt.ylabel('Global Sales (Millions)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
correlation = publisher_stats['Game_Count'].corr(publisher_stats['Global_Sales'])
print(f"Correlation between the number of games and sales: {correlation:.2f}")
```

A correlation of 0.80 indicates a strong positive relationship between the number of games published and global sales. In other words:
The more games a distributor releases, the higher its total sales volume tends to be.
Of course, this doesn't necessarily mean quality; a distributor can release many mediocre games and still sell a lot. But this data clearly shows how volume influences commercial success.

How much does each distributor depend on its most profitable genre, that is, whether it is more diversified or focuses heavily on a single style? Let's take the top 15 best-selling brands and do this analysis:

```python
# Passo 1: Vendas totais por publisher
publisher_total_sales = df.groupby('Publisher')['Global_Sales'].sum().reset_index()
publisher_total_sales.rename(columns={'Global_Sales': 'Total_Sales'}, inplace=True)

# Passo 2: Selecionar top 15 publishers com mais vendas
top15_publishers = publisher_total_sales.sort_values(by='Total_Sales', ascending=False).head(15)

# Passo 3: Obter o gênero mais lucrativo dessas top publishers
genre_sales_by_publisher = df.groupby(['Publisher', 'Genre'])['Global_Sales'].sum().reset_index()
top_genre_by_publisher = genre_sales_by_publisher.sort_values('Global_Sales', ascending=False).drop_duplicates('Publisher')

# Passo 4: Filtrar apenas os gêneros das top 15 publishers
top_genre_filtered = top_genre_by_publisher[top_genre_by_publisher['Publisher'].isin(top15_publishers['Publisher'])]

# Passo 5: Juntar com as vendas totais e calcular a proporção
top15_genre_dependency = top_genre_filtered.merge(top15_publishers, on='Publisher')
top15_genre_dependency['Top_Genre_Percent'] = (top15_genre_dependency['Global_Sales'] / top15_genre_dependency['Total_Sales']) * 100

# Ordenar para visualização
top15_genre_dependency = top15_genre_dependency.sort_values('Top_Genre_Percent', ascending=False)

# Exibir a tabela
top15_genre_dependency[['Publisher', 'Genre', 'Global_Sales', 'Total_Sales', 'Top_Genre_Percent']]

```

## Conclusion: Gender Dependency in Major Distributors

An analysis of the **15 distributors with the highest global sales volume** revealed interesting patterns regarding the **portfolio diversification** and **commercial strategies** adopted by each:

### 1. Dominant Genres
- Each distributor has a **main genre** that accounts for the majority of its revenue.
- Genres such as **Action**, **Sports**, and **Platform** were the most common among the top sellers.

### 2. Degree of Dependency
- Some publishers such as **Nintendo** and **Electronic Arts (EA)** demonstrate an **high dependence** on a single genre:
  - **Nintendo** has a large portion of its sales concentrated on the **Platform** (ex: Mario games).
  - **EA** It relies heavily on the **Sports** genre, driven by franchises like FIFA and Madden.
- Others, such as **Ubisoft** or **Sony**, show a more balanced distribution of sales across multiple genres, suggesting a more diversified strategy.

### 3. Commercial Strategy
- Distributors with high dependency tend to focus on **established franchises** and **recurring product launches**.
- Diversified companies, on the other hand, demonstrate market flexibility and greater resilience to changing trends



### Relevant Insights
- The strong **correlation between the number of games published and total sales (r = 0.80)** shows that **volume matters**, but the **right genre** can be even more decisive.
- Focusing on a single genre **can generate large profits**, but it also **increases risk** in scenarios of market saturation or changing demand.
