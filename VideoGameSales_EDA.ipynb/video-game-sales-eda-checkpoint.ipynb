# Análise Exploratória da Base de Video Games Sales 

Neste notebook, vamos explorar a base de dados de vendas de video games, identificando padrões de vendas por ano, gênero, plataforma e região.
Dataset importado do Kaggle: https://www.kaggle.com/datasets/sidtwr/videogames-sales-dataset/data

Inicializo o tratamento, importando as bibliotecas Pandas, Matplotlib e Seaborn e ajustando a configuração estética dos futuros gráficos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações estéticas
sns.set(style="whitegrid")
%matplotlib inline
```

Leio o dataset e busco vejo as primeiras 5 linhas:

```python
df = pd.read_csv("vgsales.csv") 
df.head()
```

Agora vamos verificar o tipo de dado que possuímos em cada coluna:

```python
df.info()
```

Verificamos então que temos 4 variáveis do tipo Objeto (Nome do jogo(Name), plataforma que o abriga(Plataform), Gênero do jogo(Genre), proprietário(Publisher)) 1 váriavel tipo inteira (Rank de vendas (Rank)) e 6 tipo flutuante(Ano de publicação(Year), vendas na América do norte(NA_Sales), vendas na Europa(EU_Sales), vendas no japão(JP_Sales), vendas no restante do mundo(Other_Sales) e vendas globais(Global_Sales)). Temos 16598 linhas e 11 Colunas no total. 

Posteriormente, começamos o tratamento de dados verificando os espaços vazios:

```python
df.isnull().sum()
```

Verificamos então que possuimos 271 valores faltantes em Year(ano) e 58 valores faltantes em Publisher(Distribuidora). Dessa forma irei trocar os valores nulos em Publisher por 'Unknown'. Quanto ao ano, pensei na possibilidade de colocar a média, no entanto, não é viável, dado que não reflete a realiade. Pensei em excluir as linhas, mas isso traria o problema de excluir alguns jogos e enviesária nossa análise. Portanto decidi não fazer análises utilizando o ano, talvez seria interessante essa análise em um dataset mais completo.

```python
df["Publisher"] = df["Publisher"].fillna("Unknown")
```

Agora iremos plotar os top 10 jogos mais vendidos globalmente, defini o eixo x com o nome de cada jogo, e eixo y com a quantidade de vendas globais. Defini a paleta de cores como Viridis para a coloração ficar mais intuitiva. 

```python
top10 = df.sort_values(by="Global_Sales", ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(data=top10, x="Name", y="Global_Sales", palette="viridis")
plt.title("Top 10 Jogos Mais Vendidos Globalmente")
plt.xticks(rotation=45)
plt.show()
```

Agora vamos ver se esse gráfico muda de acordo com a sua região, iniciando com a América do norte:

```python
top10 = df.sort_values(by="NA_Sales", ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(data=top10, x="Name", y="NA_Sales", palette="viridis")
plt.title("Top 10 Jogos Mais Vendidos Norte América")
plt.xticks(rotation=45)
plt.show()
```

Comparando os dois gráficos verificamos que Wii Sports e Super Mario Bros continuam na liderança como mais vendidos. O restante muda ligeramente de posição. 
Vamos analisar agora na Europa:

```python
top10 = df.sort_values(by="EU_Sales", ascending=False).head(10) 
plt.figure(figsize=(12, 6))
sns.barplot(data=top10, x="Name", y="EU_Sales", palette="viridis")
plt.title("Top 10 Jogos Mais Vendidos na Europa")
plt.xticks(rotation=45)
plt.show()
```

Aqui já vemos uma mudança mais brusca e outros titulos aparecendo, como NintendoDogs e Grand Theft Auto V. Vamos deixar mais resumido esses gráficos para comparação:


```python
#definindo o nome das regiões para organizar o gráfico
regions = {
    "América do Norte": "NA_Sales",
    "Europa": "EU_Sales",
    "Japão": "JP_Sales",
    "Outras Regiões": "Other_Sales",
    "Global": "Global_Sales"
}

# Layout para os subgráficos
fig, axs = plt.subplots(3, 2, figsize=(16, 15))
fig.suptitle("Top 5 Jogos Mais Vendidos por Região", fontsize=20)
axs = axs.flatten()

# Loop nas regiões
for i, (region_name, column) in enumerate(regions.items()):
    # Pega os 5 jogos mais vendidos na região
    top_games = df[["Name", column]].sort_values(by=column, ascending=False).head(5)
    
    # Gráfico de barras
    sns.barplot(
        data=top_games,
        x=column,
        y="Name",
        palette="viridis",
        ax=axs[i]
    )
    
    axs[i].set_title(region_name, fontsize=14)
    axs[i].set_xlabel("Vendas (em milhões)")
    axs[i].set_ylabel("Nome do Jogo")

# Remove gráfico vazio se sobrar (3x2 = 6 slots e temos 5 regiões)
if len(regions) < len(axs):
    axs[-1].axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

Vemos que há uma mudança significativa nos top 5 de cada região. Vejamos agora se isso é reflexo dos gostos regionais por cada gênero:

```python

# Agrupando e filtrando novamente (caso esteja em nova célula)
publisher_stats = df.groupby('Publisher').agg({
    'Name': 'count',
    'Global_Sales': 'sum'
}).rename(columns={'Name': 'Game_Count'})

publisher_stats = publisher_stats[publisher_stats['Game_Count'] > 1]
top_publishers_stats = publisher_stats.sort_values(by='Global_Sales', ascending=False).head(15)

# Criando gráfico com linha de regressão
plt.figure(figsize=(12, 8))
sns.regplot(
    data=top_publishers_stats,
    x='Game_Count',
    y='Global_Sales',
    scatter=True,
    ci=None,
    line_kws={"color": "red", "linewidth": 1.5}
)

# Adicionando os nomes das publishers
for publisher, row in top_publishers_stats.iterrows():
    plt.text(row['Game_Count'] + 0.3, row['Global_Sales'], publisher, fontsize=9)

plt.title('Top 15 Publishers: Quantidade de Jogos vs. Vendas Globais (com linha de tendência)')
plt.xlabel('Quantidade de Jogos Publicados')
plt.ylabel('Vendas Globais (milhões)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
correlation = publisher_stats['Game_Count'].corr(publisher_stats['Global_Sales'])
print(f"Correlação entre quantidade de jogos e vendas: {correlation:.2f}")
```

Uma correlação de 0.80 indica uma forte relação positiva entre a quantidade de jogos publicados e as vendas globais. Ou seja:
Quanto mais jogos uma Distribuidora lança, maior tende a ser seu volume total de vendas.
Claro que isso não significa necessariamente qualidade, uma distribuidora pode lançar muitos jogos medianos e ainda assim vender bastante. Mas esse dado mostra bem como o volume influência no sucesso comercial.

O quanto cada distribuidora depende do seu gênero mais lucrativo, ou seja, se ela é mais diversificada ou se aposta pesado em um único estilo. Vamos pegar as top 15 que mais vendem e fazer essa análise:

```python
# Passo 1: Vendas totais por publisher
publisher_total_sales = df.groupby('Publisher')['Global_Sales'].sum().reset_index()
publisher_total_sales.rename(columns={'Global_Sales': 'Total_Sales'}, inplace=True)

# Passo 2: Selecionar top 15 publishers com mais vendas
top15_publishers = publisher_total_sales.sort_values(by='Total_Sales', ascending=False).head(15)

# Passo 3: Obter o gênero mais lucrativo dessas top publishers
genre_sales_by_publisher = df.groupby(['Publisher', 'Genre'])['Global_Sales'].sum().reset_index()
top_genre_by_publisher = genre_sales_by_publisher.sort_values('Global_Sales', ascending=False).drop_duplicates('Publisher')

# Passo 4: Filtrar apenas os gêneros das top 15 publishers
top_genre_filtered = top_genre_by_publisher[top_genre_by_publisher['Publisher'].isin(top15_publishers['Publisher'])]

# Passo 5: Juntar com as vendas totais e calcular a proporção
top15_genre_dependency = top_genre_filtered.merge(top15_publishers, on='Publisher')
top15_genre_dependency['Top_Genre_Percent'] = (top15_genre_dependency['Global_Sales'] / top15_genre_dependency['Total_Sales']) * 100

# Ordenar para visualização
top15_genre_dependency = top15_genre_dependency.sort_values('Top_Genre_Percent', ascending=False)

# Exibir a tabela
top15_genre_dependency[['Publisher', 'Genre', 'Global_Sales', 'Total_Sales', 'Top_Genre_Percent']]

```

## Conclusão: Dependência de Gênero nas Principais diatribuidoras

A análise das **15 distribuidoras com maior volume de vendas globais** revelou padrões interessantes sobre a **diversificação de portfólio** e **estratégias comerciais** adotadas por cada uma:

### 1. Gêneros Dominantes
- Cada distribuidora possui um **gênero principal** responsável pela maior parte de sua receita.
- Gêneros como **Ação**, **Esportes** e **Plataformas** foram os mais recorrentes entre os líderes de venda.

### 2. Grau de Dependência
- Algumas publishers como **Nintendo** e **Electronic Arts (EA)** demonstram **alta dependência** de um único gênero:
  - **Nintendo** tem grande parte de suas vendas concentradas em **Plataforma** (ex: jogos do Mario).
  - **EA** depende fortemente do gênero **Esportes**, impulsionado por franquias como FIFA e Madden.
- Outras, como **Ubisoft** ou **Sony**, apresentam uma distribuição de vendas mais equilibrada entre múltiplos gêneros, sugerindo uma estratégia mais diversificada.

### 3. Estratégia Comercial
- Distribuidoras com alta dependência tendem a apostar em **franquias consolidadas** e **lançamentos recorrentes**.
- Já as diversificadas demonstram flexibilidade de mercado e maior resiliência a mudanças de tendência.



### Insights Relevantes
- A forte **correlação entre quantidade de jogos publicados e vendas totais (r = 0.80)** mostra que **volume importa**, mas o **gênero certo** pode ser ainda mais decisivo.
- Focar em um único gênero **pode gerar grandes lucros**, mas também **aumenta o risco** em cenários de saturação de mercado ou mudança de demanda.
